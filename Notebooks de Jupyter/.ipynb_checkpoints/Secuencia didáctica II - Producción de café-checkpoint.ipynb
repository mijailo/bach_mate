```python
nq=9 #Número de preguntas
cuestionario={i:{'P':'','R':''} for i in range(1,1+nq)}
for i in cuestionario.keys():
    cuestionario[i]['P']=input('(Escribe aquí la pregunta {}.)'.format(i))
```

```python
p=dict()
for preg in cuestionario.keys():
    p[preg]=widgets.Textarea(
        value='',
        placeholder='(Escribe aquí tu respuesta)',
        description='R:',
        disabled=False
    )
    display(widgets.HBox([widgets.Label(value=cuestionario[preg]['P']), p[preg]]))
```

```python
for k in cuestionario.keys():
    cuestionario[k]['R']=p[k].value
    
cuestionario
```
